```python
mh_df.isnull()
mh_df.isnull().sum()
#There are multiple columns with missing data
```

```python
bpm_mode = mh_df['BPM'].mode()[0]
mh_df['BPM']=mh_df['BPM'].fillna(bpm_mode)
mh_df.isnull().sum()
mh_df_cleaned = mh_df.dropna(axis = 0, how= 'any')
mh_df_cleaned.isnull().sum()

mh_df_cleaned.to_csv('cleaned_mxmh_survey_results.csv', index = False)
```

## Irregular Data

Detect outliers in your dataset and handle them as needed. Use code comments to make notes about your thought process.

```python

Q1 = mh_df['BPM'].quantile(0.25)
Q3 = mh_df['BPM'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(Q1)
print(Q3)
print(IQR)
print(lower_bound)
print(upper_bound)
```
